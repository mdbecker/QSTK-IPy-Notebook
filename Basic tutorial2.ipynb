```python
# QSTK Imports
import QSTK.qstkutil.qsdateutil as du
import QSTK.qstkutil.tsutil as tsu
import QSTK.qstkutil.DataAccess as da

# Third Party Imports
import datetime as dt
import pandas as pd
import numpy as np
```

```python
figsize(20, 10)
```

```python
# This is the location of the QSTK directory relative to the path where you ran IPython Notebook
QSTK_DIR = "QSTK-0.2.5"
```

```python
# Reading the csv file.
na_data = np.loadtxt('{0}/Examples/Basic/example-data.csv'.format(QSTK_DIR), delimiter=',', skiprows=1)
na_price = na_data[:, 3:]  # Default np.loadtxt datatype is float.
na_dates = np.int_(na_data[:, 0:3])  # Dates should be int
ls_symbols = ['$SPX', 'XOM', 'GOOG', 'GLD']
```

```python
# Printing the first 5 rows
print "First 5 rows of Price Data:"
print na_price[:5, :]
print
print "First 5 rows of Dates:"
print na_dates[:5, :]
```

```python
# Creating the timestamps from dates read
ldt_timestamps = []
for i in range(0, na_dates.shape[0]):
    ldt_timestamps.append(dt.date(na_dates[i, 0],
                    na_dates[i, 1], na_dates[i, 2]))
    
# Plotting the prices with x-axis=timestamps
plot(ldt_timestamps, na_price)
legend(ls_symbols)
ylabel('Adjusted Close')
xlabel('Date')
```

```python
# Normalizing the prices to start at 1 and see relative returns
na_normalized_price = na_price / na_price[0, :]

# Plotting the prices with x-axis=timestamps
plot(ldt_timestamps, na_normalized_price)
legend(ls_symbols)
ylabel('Normalized Close')
xlabel('Date')
```

```python
# Copy the normalized prices to a new ndarry to find returns.
na_rets = na_normalized_price.copy()

# Calculate the daily returns of the prices. (Inplace calculation)
tsu.returnize0(na_rets)

# Plotting the plot of daily returns
plot(ldt_timestamps[0:50], na_rets[0:50, 0])  # $SPX 50 days
plot(ldt_timestamps[0:50], na_rets[0:50, 1])  # XOM 50 days
axhline(y=0, color='r')
legend(['$SPX', 'XOM'])
ylabel('Daily Returns')
xlabel('Date')
```

```python
# Plotting the scatter plot of daily returns between XOM VS $SPX
scatter(na_rets[:, 0], na_rets[:, 1], c='blue')
ylabel('XOM')
xlabel('$SPX')
```

```python
# Plotting the scatter plot of daily returns between $SPX VS GLD
scatter(na_rets[:, 0], na_rets[:, 3], c='blue')  # $SPX v GLD
ylabel('GLD')
xlabel('$SPX')
```
